## Components of Time Series

<p>A Time-Series has the following three important components</p>

<ol>
 <li> Trend </li>
 <li> Seasonality </li>
 <li> Random Noise </li>
</ol>

### 1) Trend

<p>Refres to a long term variation, where the output variable tends to increase or descrease over a long period of time. </p>

<img src="trend.png">

### 2) Seasonality

<p>Refers to short term variations, there variations are periodic and regular in nature.</p>

<img src="seasonality.png">

### 3) Random Noise / Irregularity

<p>Random fluctuations in time series, also known as noise or residual.</p>

## Mathematical Representation of Time Series

<p> Time Series can be represented using two mathematical models</p>

<ol>
  <li>Additive</li>
  <li>Multiplicative</li>
</ol>

### Additive

<p> Time series output is represented as addition of components. </p>

<p><b>Y = Level + Trend + Seasonality + Noise</b></p>

<p> <b> Level </b> is average/mean value of time series data </p> 

<p> Additive model is useful when seasonal variation is constant over time and do not depend on the trend </p>


<img src="add.png">

### Multiplicative

<p> Time series output is represented as multiplication of components. </p>

<p><b>Y = Level * Trend * Seasonality * Noise</b></p>

<p> Multiplicative model is useful when seasonal variation is not constant and its frequency changes with trend </p>

<img src="time.png">

## Classification of Time Series

<p><b>1) Discrete Time Series</b> - when variable of study is recorded on daily or monthly basis. Here time is in discrete form.</p>
<p> Example - Daily Stock Price </p>
<p><b>2) Continuous Time Series</b> - when variable of study is recorded at every second,minute or hour. Here time is in continuous form.</p>
<p> Example - Hourly Temperature </p>
<p><b>3) Univariate Time Series</b> -when there is only one varibale of study.</p>
<p> Example - Daily Stock Price </p>

<p><b>4) Multivariate Time Series</b> -when there are more than one varibales of study.</p>
<p> Example - Hourly Temperature and Humidity </p>


### Stationary vs Non-Stationary

<p> A Time series data is said to be <b>stationary</b> if it fulfills the following criteria </p>

<ol>
    <li>Constant mean</li>
    <li>Constant variance</li>
    <li>No seasonality</li>
</ol>

<p> If these criteria are not fulfilled, the time series data is <b>non-stationary</b></p>

## Partial Auto Correlation Function (PACF) 

<p>The effect of autocorrelation at smaller lags will have an influence on the estimate of autocorrelation at longer lags</p> 

<p>A strong lag one, can cause an autocorrelation with lag three</p>

<p>The Partial autocorrelation function (PACF) removes the effect of shorter lag autocorrelation from the correlation estimate at longer lags</p>

```python
from statsmodels.graphics.tsaplots import plot_pacf

plot_pacf(tesla['Close'], lags=20)

plt.show()
```

## LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN, LSTM

from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import mean_squared_error

import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
apple = yf.download('AAPL','2018-01-27', '2020-01-27')
```

```python
apple.head()
```

```python
plt.plot(apple["Close"])
```

```python
apple.shape
```

```python
train = apple["Close"][:352]
test = apple["Close"][352:]
```

```python
train.shape
```

```python
test.shape
```

```python
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(np.array(train).reshape(-1,1))
test_scaled = scaler.transform(np.array(test).reshape(-1,1))
```

```python
#convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        sample = dataset[i:(i+time_step), 0]    
        X.append(sample)
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)
```

```python
time_step = 50
X_train, y_train = create_dataset(train_scaled, time_step)
X_test, y_test = create_dataset(test_scaled, time_step)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# reshape into (samples, time steps, features) 
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test  = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
model = Sequential()
model.add(SimpleRNN(32, return_sequences=True, input_shape=(time_step,1)))
model.add(SimpleRNN(32))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.fit(X_train,y_train, epochs=20, batch_size=32)
```

```python
y_pred = model.predict(X_test)
mean_squared_error(y_test,y_pred)
```

```python
plt.figure()
plt.plot(scaler.inverse_transform(np.array(y_test).reshape(-1,1)))
plt.plot(scaler.inverse_transform(y_pred))
plt.show()
```

```python
model = Sequential()
model.add(LSTM(32, return_sequences=True, input_shape=(time_step,1)))
model.add(LSTM(32, return_sequences=True))
model.add(LSTM(32))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.fit(X_train,y_train, epochs=50, batch_size=32)
```

```python
y_pred = model.predict(X_test)
mean_squared_error(y_test,y_pred)
```

```python
plt.figure()
plt.plot(scaler.inverse_transform(np.array(y_test).reshape(-1,1)))
plt.plot(scaler.inverse_transform(y_pred))
plt.show()
```

```python
# next 50 days

days = 50
last_input = X_test[-1]
last_output = y_pred[-1]
y_forecast = []

for i in range(1,days+1):
    last_input = np.append(last_input[1:], last_output)
    last_output = model.predict(last_input.reshape(1,50,1))
    y_forecast.append(last_output[0][0])
    
```

```python
all_predictions = np.append(y_pred,y_forecast)
```

```python
plt.figure()
plt.plot(scaler.inverse_transform(np.array(y_test).reshape(-1,1)))
plt.plot(scaler.inverse_transform(all_predictions.reshape(-1,1)))
plt.show()
```
